```python
from PIL import Image

background = Image.new('RGBA', (701, 731), (255, 0, 0, 1))
background
```

```python
def brightness(image):
   im = image.convert('L')
   stat = ImageStat.Stat(im)
   return stat.rms[0]

#side -  top  bottom
#month_png - 
def append_calendar(img, background, side, month_image_path):
    month_image = Image.open(month_image_path, 'r').convert("RGBA")
    month_image_w, month_image_h = month_image.size
    width, height = img.size
    bg_w, bg_h = background.size
    offset_w = (bg_w - width) // 2
    offset_h = (bg_h - height) // 2
    
    if brightness(background) < 100:
        r,g,b,a = month_image.split()
        rgb_month_image = Image.merge('RGB', (r,g,b))
        month_image_inverted = PIL.ImageOps.invert(rgb_month_image)
        r2,g2,b2 = month_image_inverted.split()
        month_image = Image.merge('RGBA', (r2,g2,b2,a))
    if side == 'top':
        background.paste(month_image, ((bg_w - month_image_w) // 2, (offset_h - month_image_h) // 2), mask=month_image)
    else:
        background.paste(month_image, ((bg_w - month_image_w) // 2, (offset_h + height) + ((offset_h - month_image_h)//2)), mask=month_image)
    return background





background = Image.new('RGBA', (701, 731), (255, 255, 150, 150))


bg_w, bg_h = background.size
img.thumbnail((bg_w, bg_h))
width, height = img.size

offset_w = (bg_w - width) // 2
offset_h = (bg_h - height) // 2

offset = (offset_w, offset_h)

background.paste(img, offset)


#        
background = append_calendar(img, background, 'top', "top_6.png")
background = append_calendar(img, background, 'bottom', "bottom_6.png")
#     
#background = append_calendar(img, background, 'top', "january.png")

background
```
